```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stt
```

```python
simulation = 1_000_000
mu, sigma = 0, 1
alpha = 0.01
x = 1
X = np.random.normal(mu, sigma, simulation)

plt.figure(figsize=(8, 4))
plt.hist(X, bins=100, density=True, alpha=0.6, color='g', label='Sample Data')

xmin, xmax = plt.xlim()
x_range = np.linspace(xmin, xmax, 100)
pdf = stt.norm.pdf(x_range, mu, sigma)
plt.plot(x_range, pdf, 'k', label='Normal Distribution')

plt.axvline(-x, color='r', linestyle='--', linewidth=2, label='-x')
plt.axvline(x, color='r', linestyle='--', linewidth=2, label='x')

plt.legend()
plt.show()

p_sample = np.logical_and(-x < X, X < x).sum() / simulation
print(f'Sample proportion: {p_sample}')

f = stt.norm(mu, sigma)
p_population = f.cdf(x) - f.cdf(-x)
print(f'Population proportion: {p_population}')

print(f'Epsilon: {np.abs(p_sample - p_population)}')

print(f'Maximum value in sample data: {X.max()}')
```
